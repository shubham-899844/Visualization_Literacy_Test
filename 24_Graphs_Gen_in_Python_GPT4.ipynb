# HeatMap or Area Chart
Visualize the frequency of unusual activities (like pacing or falling) across different camera zones over a 24-hour period.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulated data: frequency of events in each zone for every hour
zones = ['Front Gate', 'Backyard', 'Living Room', 'Hallway', 'Driveway']
hours = list(range(24))

# Create a DataFrame with random frequencies for each zone per hour
np.random.seed(42)  # For reproducibility
data = {
    zone: np.random.poisson(lam=np.random.randint(2, 6), size=24) for zone in zones
}
df = pd.DataFrame(data, index=hours)
df.index.name = "Hour"

# Plotting the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.T, cmap='YlOrRd', annot=True, fmt="d", cbar_kws={'label': 'Frequency'})
plt.title("Frequency of Unusual Activities by Zone Over 24 Hours")
plt.xlabel("Hour of Day")
plt.ylabel("Camera Zone")
plt.tight_layout()
plt.show()
```

# Box Plot
Compare the latency of event detection for three ViT model variants across different edge hardware setups.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulated models and hardware
models = ['ViT-Base', 'ViT-Large', 'ViT-EdgeOptimized']
hardware_setups = ['Raspberry Pi', 'NVIDIA Jetson', 'Qualcomm Snapdragon']

# Simulate latency data (in milliseconds)
np.random.seed(42)
data = []

for model in models:
    for hardware in hardware_setups:
        latencies = np.random.normal(loc=np.random.randint(30, 100), scale=5, size=30)  # 30 samples
        for latency in latencies:
            data.append({'Model': model, 'Hardware': hardware, 'Latency (ms)': latency})

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the boxplot
plt.figure(figsize=(14, 7))
sns.boxplot(x="Hardware", y="Latency (ms)", hue="Model", data=df, palette="Set3")
plt.title("Latency Comparison of ViT Model Variants Across Edge Hardware Setups")
plt.xlabel("Edge Hardware Setup")
plt.ylabel("Latency (milliseconds)")
plt.legend(title="ViT Model")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Bar Chart
Compare the average frame-processing time per camera when using different ViT architectures.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulated cameras and ViT architectures
cameras = ['Front Gate', 'Backyard', 'Living Room', 'Garage', 'Hallway']
vit_models = ['ViT-Base', 'ViT-Large', 'ViT-EdgeOptimized']

# Simulate average frame-processing time (in milliseconds)
np.random.seed(42)
data = {
    'Camera': [],
    'ViT Model': [],
    'Avg Frame Processing Time (ms)': []
}

for camera in cameras:
    for model in vit_models:
        time = np.random.uniform(20, 120)  # Random time between 20ms and 120ms
        data['Camera'].append(camera)
        data['ViT Model'].append(model)
        data['Avg Frame Processing Time (ms)'].append(round(time, 2))

# Create DataFrame
df = pd.DataFrame(data)

# Pivot to make Camera on X-axis, Models as groups
df_pivot = df.pivot(index='Camera', columns='ViT Model', values='Avg Frame Processing Time (ms)')

# Plotting the grouped bar chart
ax = df_pivot.plot(kind='bar', figsize=(12, 7), colormap='tab20')

plt.title("Average Frame Processing Time per Camera Across ViT Architectures")
plt.xlabel("Camera")
plt.ylabel("Avg Frame Processing Time (ms)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="ViT Model")
plt.tight_layout()
plt.show()
```

# Scatter Plot with color encoding
Visualize correlation between lighting conditions and failure rates in activity recognition.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulated data
np.random.seed(42)
n_samples = 150

# Lighting in Lux (0 = dark, 10000 = bright)
lighting = np.random.uniform(50, 10000, n_samples)
# Failure rate increases in low light
failure_rate = 30 - (lighting / 400) + np.random.normal(0, 3, n_samples)
failure_rate = np.clip(failure_rate, 0, 30)  # Clamp between 0 and 30%

# Randomly assign activity type
activity_types = np.random.choice(['Loitering', 'Falling', 'Intrusion', 'Running'], n_samples)

# Create DataFrame
df = pd.DataFrame({
    'Lighting (lux)': lighting,
    'Failure Rate (%)': failure_rate,
    'Activity Type': activity_types
})

# Plotting
plt.figure(figsize=(12, 7))
scatter = sns.scatterplot(
    data=df,
    x='Lighting (lux)', 
    y='Failure Rate (%)',
    hue='Activity Type',
    palette='tab10',
    style='Activity Type',
    s=100  # size of points
)

plt.title('Correlation between Lighting Conditions and Failure Rates in Activity Recognition')
plt.xlabel('Lighting Condition (lux)')
plt.ylabel('Failure Rate (%)')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend(title='Activity Type')
plt.tight_layout()
plt.show()
```

# Radar Chart
Compare how well the model detects falls vs. lying down using attention weight scores.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the categories for comparison
categories = ['Sensitivity', 'Specificity', 'Precision', 'Recall', 'F1 Score']

# Define the scores for fall detection and lying down detection
fall_detection_scores = [0.85, 0.90, 0.88, 0.87, 0.89]  # Example values for fall detection
lying_down_scores = [0.80, 0.95, 0.82, 0.79, 0.81]  # Example values for lying down detection

# Number of variables (categories)
num_vars = len(categories)

# Compute the angle for each category (evenly spaced)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Make the radar chart "close the loop" by repeating the first value
fall_detection_scores += fall_detection_scores[:1]
lying_down_scores += lying_down_scores[:1]
angles += angles[:1]

# Create the plot
fig, ax = plt.subplots(figsize=(6, 6), dpi=100, subplot_kw=dict(polar=True))

# Plot the data for fall detection
ax.plot(angles, fall_detection_scores, linewidth=2, linestyle='solid', label='Fall Detection', color='r')
ax.fill(angles, fall_detection_scores, color='r', alpha=0.25)

# Plot the data for lying down detection
ax.plot(angles, lying_down_scores, linewidth=2, linestyle='solid', label='Lying Down Detection', color='b')
ax.fill(angles, lying_down_scores, color='b', alpha=0.25)

# Set the labels for each category
ax.set_yticklabels([])  # Hide the radial ticks
ax.set_xticks(angles[:-1])  # Set the angular ticks
ax.set_xticklabels(categories, fontsize=12)

# Add a legend
ax.legend(loc='upper right', fontsize=12)

# Display the plot
plt.title('Comparison of Fall Detection vs Lying Down')
```

#  Stacked Column Chart or Pie Chart)
Visualize how often human override was needed after a system-generated alert, per event category.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data: Human override counts after system-generated alerts, per event category
event_categories = ['Motion Detected', 'Unusual Activity', 'Security Alert', 'Sensor Failure']
alerts_generated = np.array([100, 150, 120, 90])  # Total alerts generated by the system
human_overrides = np.array([30, 50, 40, 20])  # Human overrides needed for each event category

# Calculating the remaining alerts after human override
remaining_alerts = alerts_generated - human_overrides

# Creating a stacked column chart
plt.bar(event_categories, remaining_alerts, label='Remaining Alerts', color='b')
plt.bar(event_categories, human_overrides, label='Human Overrides', color='r', bottom=remaining_alerts)

# Adding labels and title
plt.xlabel('Event Category', fontsize=12)
plt.ylabel('Count of Alerts', fontsize=12)
plt.title('Human Override Needed After System-Generated Alerts', fontsize=14)

# Adding legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
