```python
import matplotlib.pyplot as plt
import random

# Data for metro systems
cities = [
    "Tokyo", "Delhi", "Shanghai", "New York City", "Paris",
    "Beijing", "London", "Moscow", "Seoul", "Guangzhou", "Mexico City"
]
stations = [180, 150, 350, 420, 300, 330, 260, 230, 280, 150, 200]         # x-axis
length_km = [200, 180, 550, 340, 250, 520, 400, 320, 330, 240, 210]         # y-axis
ridership = [2.3, 2.1, 3.5, 2.0, 1.5, 3.3, 1.8, 1.6, 2.0, 1.7, 1.4]         # bubble size

# Plot bubble chart
plt.figure(figsize=(12, 8))
sizes = [r * 300 for r in ridership]
plt.scatter(stations, length_km, s=sizes, alpha=0.5)

# Annotate each city
for i in range(len(cities)):
    plt.text(stations[i] + 3, length_km[i], cities[i], fontsize=9)

# Labels and title
plt.xlabel("Stations", fontsize=12, weight='bold')
plt.ylabel("Total Length", fontsize=12, weight='bold')
plt.title("Metro Systems of the World", fontsize=14, weight='bold')

# Question pool
questions = [
    "Q: The metro system in Shanghai has more ridership than in Beijing.\n   A. True   B. False",
    "Q: Which metro has the highest number of stations?\n   A. New York City   B. London   C. Shanghai   D. Tokyo",
    "Q: Which city has a longer network: Beijing or Paris?\n   A. Beijing   B. Paris"
]
question = random.choice(questions)

# Add question as text below the chart
plt.text(50, -50, question, fontsize=11, bbox=dict(facecolor='lightgray', edgecolor='black'))

# Adjust layout and show
plt.ylim(150, 600)
plt.xlim(120, 470)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import pandas as pd
import random

# Sample unemployment rate data (in percentage as decimals)
data = {
    'State': ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN',
              'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV',
              'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN',
              'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'],
    'Unemployment': [
        0.061, 0.068, 0.064, 0.059, 0.065, 0.037, 0.052, 0.050, 0.053, 0.058, 0.042, 0.037, 0.060, 0.048,
        0.036, 0.041, 0.051, 0.062, 0.044, 0.047, 0.045, 0.056, 0.037, 0.067, 0.050, 0.039, 0.031, 0.066,
        0.038, 0.055, 0.066, 0.052, 0.057, 0.030, 0.051, 0.043, 0.057, 0.051, 0.061, 0.055, 0.032, 0.058,
        0.045, 0.035, 0.032, 0.045, 0.054, 0.062, 0.047, 0.037
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create choropleth map
fig = px.choropleth(
    df,
    locations='State',
    locationmode="USA-states",
    color='Unemployment',
    scope="usa",
    color_continuous_scale='Blues',
    labels={'Unemployment': 'Unemployment Rate'},
    title='Unemployment Rates for States in 2015'
)

# Add a random question as annotation
questions = [
    "Q: In 2015, was unemployment in WA higher than in WI?",
    "Q: Which state had one of the highest unemployment rates?\n   A. NM   B. IA   C. SD",
    "Q: Was the Midwest generally lighter or darker in shade?"
]
question = random.choice(questions)

fig.add_annotation(
    text=question,
    x=0.5,
    y=-0.15,
    showarrow=False,
    xref="paper",
    yref="paper",
    font=dict(size=14),
    align="center",
    bgcolor="lightgray",
    bordercolor="black"
)

# Show figure
fig.show()
```
